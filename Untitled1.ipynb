```python
from tkinter import*
from tkinter import ttk,messagebox
from PIL import Image,ImageTk
import pymysql

class Hostel:
    
    def __init__(self,wind1):
        self.wind1=wind1
        self.wind1.title("Registration window".center(400))
        self.wind1.geometry("1350x700+0+0")
        #----Background   Image----#
        self.bkgrnd=ImageTk.PhotoImage(file="images.png")
        bkgrnd=Label(self.wind1,image=self.bkgrnd).place(x=0,y=0,relwidth=1,relheight=1)
        
        #----Logo Image-----#
        self.logo=ImageTk.PhotoImage(file="images1side.png")
        logo=Label(self.wind1,image=self.logo).place(x=80,y=100,width=400,height=500)

        #-----DAV LOGO TEXT-----#
        self.lbl1=Label(self.wind1,text="DAV UNIVERSITY",font=("Bok Antiqua",25,"bold"),fg="black",compound=BOTTOM,bg="white")
        self.lbl1.place(x=90,y=100,height=50,width=350)

        frame1=Frame(self.wind1,bg="white")
        frame1.place(x=480,y=100,width=700,height=500)

        #----Login text----#
        title=Label(frame1,text="LOGIN HERE",font=("times new roman",30,"bold"),bg="white",fg="#2744AB").place(x=80,y=50)

        #-----Login Email-----#
        self.l_email=Label(frame1,text="Email",font=("times new roman",20,"bold"),bg="white",fg="gray")
        self.l_email.place(x=80,y=120)

        self.txt_l_email=Entry(frame1,font=("times new roman",15),bg="lightgray")
        self.txt_l_email.place(x=80,y=160,width=400,height=25)

        #-----Login password-----#
        self.l_password=Label(frame1,text="Password",font=("times new roman",20,"bold"),bg="white",fg="gray")
        self.l_password.place(x=80,y=220)

        self.txt_l_password=Entry(frame1,font=("times new roman",15),bg="lightgray")
        self.txt_l_password.place(x=80,y=260,width=400,height=25)

        self.btn_reg=Button(frame1,text="Register New Account?",command=self.register_page,font=("times new roman",18,"bold"),bg="white",bd=0,fg="#D42569",cursor="hand2")
        self.btn_reg.place(x=80,y=300)

        self.btn_login=Button(frame1,text="Login",font=("times new roman",18,"bold"),command=self.management,bg="white",bd=0,fg="#D42569",cursor="hand2")
        self.btn_login.place(x=80,y=350,)        
        

    def register_page(self):
        self.wind1.destroy()
        self.wind2=Tk()
        self.wind2.title("Login Window".center(400))
        self.wind2.geometry("1350x700+0+0")
        #----Background   Image-----#
        self.backgrnd=ImageTk.PhotoImage(file="images/backgb2.png")
        backgrnd=Label(self.wind2,image=self.backgrnd).place(x=0,y=0,relwidth=1,relheight=1)
        
        #----Logo Image-----#
        self.logo=ImageTk.PhotoImage(file="images/1side.png")
        centr=Label(self.wind2,image=self.logo).place(x=80,y=100,width=400,height=500)

        
        #-----DAV LOGO TEXT-----#
        self.lbl2=Label(self.wind2,text="DAV UNIVERSITY",font=("Bok Antiqua",25,"bold"),fg="black",compound=BOTTOM,bg="white")
        self.lbl2.place(x=90,y=100,height=50,width=350)

        #----Register frame2-----#
        frame2=Frame(self.wind2,bg="white")
        frame2.place(x=480,y=100,width=700,height=500)

        title=Label(frame2,text="REGISTER HERE",font=("times new roman",20,"bold"),bg="white",fg="green").place(x=50,y=30)
        
        self.f_name=Label(frame2,text="First Name",font=("times new roman",15,"bold"),bg="white",fg="gray")
        self.f_name.place(x=50,y=100)
        self.txt_fname=Entry(frame2,font=("times new roman",15),bg="lightgray")
        self.txt_fname.place(x=50,y=130,width=250)

        self.l_name=Label(frame2,text="Last Name",font=("times new roman",15,"bold"),bg="white",fg="gray")
        self.l_name.place(x=370,y=100)
        self.txt_lname=Entry(frame2,font=("times new roman",15),bg='lightgray')
        self.txt_lname.place(x=370,y=130,width=250)

        #2ROW
        self.contact=Label(frame2,text="Contact No.",font=("times new roman",15,"bold"),bg="white",fg="gray")
        self.contact.place(x=50,y=170)
        self.txt_contact=Entry(frame2,font=("times new roman",15),bg='lightgray')
        self.txt_contact.place(x=50,y=200,width=250)


        self.email=Label(frame2,text="Email",font=("times new roman",15,"bold"),bg="white",fg="gray")
        self.email.place(x=370,y=170)
        self.txt_email=Entry(frame2,font=("times new roman",15),bg='lightgray')
        self.txt_email.place(x=370,y=200,width=250)

        #3ROW
        self.password=Label(frame2,text="Password",font=("times new roman",15,"bold"),bg="white",fg="gray")
        self.password.place(x=50,y=240)
        self.txt_password=Entry(frame2,font=("times new roman",15),bg='lightgray')
        self.txt_password.place(x=50,y=270,width=250)

        self.cpassword=Label(frame2,text="Confirm   Password",font=("times new roman",15,"bold"),bg="white",fg="gray")
        self.cpassword.place(x=370,y=240)
        self.txt_cpassword=Entry(frame2,font=("times new roman",15),bg='lightgray')
        self.txt_cpassword.place(x=370,y=270,width=250)

        #4ROW--terms--
        self.btn_img1=ImageTk.PhotoImage(file="images/rbutton.jpg")
        self.btn_register=Button(frame2,image=self.btn_img1,bd=0,cursor="hand2",command=self.register_data)
        self.btn_register.place(x=50,y=370)

        #self.btn_img2=ImageTk.PhotoImage(file="images/sbutton.png")
        #self.btn_signup=Button(self.wind2,image=self.btn_img2,bd=0,cursor="hand2")
        #self.btn_signup.place(x=180,y=500)


    def management(self):
        self.wind1.destroy()
        self.wind3=Tk()
        self.wind3.title("Hostel Management System".center(400))
        self.wind3.geometry("1350x700+0+0")

        #----Background   Image-----#
        self.bckgrnd=ImageTk.PhotoImage(file="images/backgb2.png")
        bckgrnd=Label(self.wind3,image=self.bckgrnd).place(x=0,y=0,relwidth=1,relheight=1)

        title=Label(self.wind3,text="North Hostel DAV University",bd=10,relief=GROOVE,font=("times new roman",40,"bold") ,fg="black")
        title.pack(side=TOP,fill=X)

        #----All Variables----#
        self.Roll_No_var=StringVar()
        self.name_var=StringVar()
        self.email_var=StringVar()
        self.rtype_var=StringVar()
        self.contact_var=StringVar()
        self.dob_var=StringVar()

        self.search_by=StringVar()
        self.search_txt=StringVar()
        
        
        #---Manage Frame---#
        Manage_Frame=Frame(self.wind3,bd=4,relief=RIDGE)
        Manage_Frame.place(x=20,y=100,width=450,height=580)

        m_title=Label(Manage_Frame,text="Manage Students",font=("times new roman",30,"bold"))
        m_title.grid(row=0,columnspan=2,pady=20)

        lbl_roll=Label(Manage_Frame,text="Roll No.",font=("times new roman",20,"bold"))
        lbl_roll.grid(row=1,column=0,padx=20,pady=10,sticky="w")

        txt_Roll=Entry(Manage_Frame,textvariable=self.Roll_No_var,font=("times new roman",15,"bold"),bd=5,relief=GROOVE)
        txt_Roll.grid(row=1,column=1,padx=20,pady=10,sticky="w")

        lbl_name=Label(Manage_Frame,text="Name",font=("times new roman",20,"bold"))
        lbl_name.grid(row=2,column=0,padx=20,pady=10,sticky="w")

        txt_Name=Entry(Manage_Frame,textvariable=self.name_var,font=("times new roman",15,"bold"),bd=5,relief=GROOVE)
        txt_Name.grid(row=2,column=1,padx=20,pady=10,sticky="w")

        lbl_email=Label(Manage_Frame,text="Email",font=("times new roman",20,"bold"))
        lbl_email.grid(row=3,column=0,padx=20,pady=10,sticky="w")

        txt_Email=Entry(Manage_Frame,textvariable=self.email_var,font=("times new roman",15,"bold"),bd=5,relief=GROOVE)
        txt_Email.grid(row=3,column=1,padx=20,pady=10,sticky="w")

        lbl_rtype=Label(Manage_Frame,text="Room Type",font=("times new roman",20,"bold"))
        lbl_rtype.grid(row=4,column=0,padx=20,pady=10,sticky="w")

        combo_type=ttk.Combobox(Manage_Frame,textvariable=self.rtype_var,font=("times new roman",14,"bold"),state='readonly')
        combo_type['values']=("1 sitter non-A.C","2 sitter non-A.C","3 sitter non-A.C","1 sitter A.C","2 sitter A.C","3 sitter A.C")
        combo_type.grid(row=4,column=1,padx=20,pady=10,sticky="w")

        lbl_contact=Label(Manage_Frame,text="Contact",font=("times new roman",20,"bold"))
        lbl_contact.grid(row=5,column=0,padx=20,pady=10,sticky="w")

        txt_Contact=Entry(Manage_Frame,textvariable=self.contact_var,font=("times new roman",15,"bold"),bd=5,relief=GROOVE)
        txt_Contact.grid(row=5,column=1,padx=20,pady=10,sticky="w")

        lbl_dob=Label(Manage_Frame,text="D.O.B",font=("times new roman",20,"bold"))
        lbl_dob.grid(row=6,column=0,padx=20,pady=10,sticky="w")

        txt_DOB=Entry(Manage_Frame,textvariable=self.dob_var,font=("times new roman",15,"bold"),bd=5,relief=GROOVE)
        txt_DOB.grid(row=6,column=1,padx=20,pady=10,sticky="w")

        lbl_address=Label(Manage_Frame,text="Address",font=("times new roman",20,"bold"))
        lbl_address.grid(row=7,column=0,padx=20,pady=10,sticky="w")

        self.txt_Address=Text(Manage_Frame,width=30,height=4,font=("",10))
        self.txt_Address.grid(row=7,column=1,padx=20,pady=10,sticky="w")

        #----button Frame----#
        btn_Frame=Frame(Manage_Frame,bd=4)
        btn_Frame.place(x=15,y=520,width=420)
        
        Add_btn=Button(btn_Frame,text="Add",width=10,command=self.add_students).grid(row=0,column=0,padx=10,pady=10)
        update_btn=Button(btn_Frame,text="Upadte",width=10,command=self.update_data).grid(row=0,column=1,padx=10,pady=10)
        delete_btn=Button(btn_Frame,text="Delete",width=10,command=self.delete_data).grid(row=0,column=2,padx=10,pady=10)
        clear_btn=Button(btn_Frame,text="Clear",width=10,command=self.clear).grid(row=0,column=3,padx=10,pady=10)

        #---Detail Frame----#
        Detail_Frame=Frame(self.wind3,bd=4,relief=RIDGE)
        Detail_Frame.place(x=500,y=100,width=800,height=580)

        lbl_Search=Label(Detail_Frame,text="Search By",font=("times new roman",20,"bold"))
        lbl_Search.grid(row=0,column=0,padx=20,pady=10,sticky="w")

        combo_search=ttk.Combobox(Detail_Frame,textvariable=self.search_by,width=10,font=("times new roman",14,"bold"),state='readonly')
        combo_search['values']=("Roll_no","Name","Contact")
        combo_search.grid(row=0,column=1,padx=20,pady=10,)

        txt_Search=Entry(Detail_Frame,textvariable=self.search_txt,width=20,font=("times new roman",10,"bold"),bd=5,relief=GROOVE)
        txt_Search.grid(row=0,column=2 ,padx=20,pady=10,sticky="w")

        search_btn=Button(Detail_Frame,text="Search",width=10,pady=5,command=self.search_data).grid(row=0,column=3,padx=10,pady=10)
        showall_btn=Button(Detail_Frame,text="Show All",width=10,pady=5,command=self.fetch_data).grid(row=0,column=4,padx=10,pady=10)

        #---Table frame----#
        Table_Frame=Frame(Detail_Frame,bd=4,relief=RIDGE)
        Table_Frame.place(x=10,y=70,width=760,height=500)

        scroll_x=Scrollbar(Table_Frame,orient=HORIZONTAL)
        scroll_y=Scrollbar(Table_Frame,orient=VERTICAL)
        self.Student_table=ttk.Treeview(Table_Frame,columns=("roll","name","email","room type","contact","dob","address"),xscrollcommand=scroll_x.set,yscrollcommand=scroll_y.set)
        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=RIGHT,fill=Y)
        scroll_x.config(command=self.Student_table.xview) 
        scroll_y.config(command=self.Student_table.yview)
        self.Student_table.heading("roll",text="Roll No.")
        self.Student_table.heading("name",text="Name")
        self.Student_table.heading("email",text="Email")
        self.Student_table.heading("room type",text="Room Type")
        self.Student_table.heading("contact",text="Contact")
        self.Student_table.heading("dob",text="D.O.B")
        self.Student_table.heading("address",text="Address")
        self.Student_table['show']='headings'
        self.Student_table.column("roll",width=100)
        self.Student_table.column("name",width=100)
        self.Student_table.column("email",width=100)
        self.Student_table.column("room type",width=100)
        self.Student_table.column("contact",width=100)
        self.Student_table.column("dob",width=100)
        self.Student_table.column("address",width=150)
        
        self.Student_table.pack(fill=BOTH,expand=1)
        self.Student_table.bind("<ButtonRelease-1>",self.get_cursor)

        self.fetch_data()
    
    
    def clear(self):
        self.txt_fname.delete(0,END)
        self.txt_lname.delete(0,END)
        self.txt_contact.delete(0,END)
        self.txt_email.delete(0,END)
        self.txt_password.delete(0,END)
        self.txt_cpassword.delete(0,END)


    def register_data(self):
        if self.txt_fname.get()=="" or self.txt_contact.get()=="" or self.txt_email.get()=="" or self.txt_password.get()=="" or self.txt_cpassword.get()=="":
            messagebox.showerror("Error","All Fields Are Required",parent=self.wind2)

        elif self.txt_password.get()!=self.txt_cpassword.get():
            messagebox.showerror("Error","Password & Confirm Password should be same",parent=self.wind2)

        else:
            try:
                con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
                cur=con.cursor()
                cur.execute("select * from register where email=%s",self.txt_email.get())
                row=cur.fetchone()
                if row!=None:
                    messagebox.showerror("Error","User already exist,Please try with another email",parent=self.wind2)
                else:
                    cur.execute("insert into register (f_name,l_name,contact,email,password) values(%s,%s,%s,%s,%s)",
                            (
                                self.txt_fname.get(),
                                self.txt_lname.get(),
                                self.txt_contact.get(),
                                self.txt_email.get(),
                                self.txt_password.get()
                            ))
                    con.commit()
                    con.close()
                    messagebox.showinfo("Success","Register Successfull",parent=self.wind2)
                    self.clear()
            except Exception as es:
                 messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.wind2)


    def login_data(self):
        if self.txt_l_email.get()=="" or self.txt_l_password.get()=="":
            messagebox.showerror("Error","All Fields Are Required",parent=self.wind1)

        else:
            try:
                con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
                cur=con.cursor()
                cur.execute("select * from register where email=%s and password=%s",(self.txt_l_email.get(),self.txt_l_password.get()))
                row=cur.fetchone()
                if row==None:
                    messagebox.showerror("Error","INVALID USER NAME AND PASSWORD",parent=self.wind1)
                else:
                    messagebox.showinfo("Success","Welcome",parent=self.wind1)
            except Exception as es:
                 messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.wind1)


    def add_students(self):
        if self.Roll_No_var.get()=="" or self.name_var.get()=="" or self.email_var.get()=="" or self.rtype_var.get()=="" or self.contact_var.get()=="" or self.dob_var.get()=="":
            messagebox.showerror("Erroe","All fields are required!!!",parent=self.wind3)
        
        else:
            con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
            cur=con.cursor()
            cur.execute("select * from students where email=%s",self.email_var.get())
            row=cur.fetchone()
            cur.execute("insert into students values(%s,%s,%s,%s,%s,%s,%s)",(self.Roll_No_var.get(),
                                                                         self.name_var.get(),
                                                                         self.email_var.get(),
                                                                         self.rtype_var.get(),
                                                                         self.contact_var.get(),
                                                                         self.dob_var.get(),
                                                                         self.txt_Address.get('1.0',END)
                                                                         ))
            con.commit()
            self.fetch_data()
            self.clear()
            con.close()
            messagebox.showinfo("Success","Record has been inserted")

    def fetch_data(self):
        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()
        cur.execute("select * from students")
        rows=cur.fetchall()
        if len(rows)!=0:
            self.Student_table.delete(*self.Student_table.get_children())
            for row in rows:
                self.Student_table.insert('',END,values=row)
            con.commit()
        con.close()

    def clear(self):
        self.Roll_No_var.set("")
        self.name_var.set("")
        self.email_var.set("")
        self.rtype_var.set("")
        self.contact_var.set("")
        self.dob_var.set("")
        self.txt_Address.delete("1.0",END)

    def get_cursor(self,ev):
        cursor_row=self.Student_table.focus()
        contents=self.Student_table.item(cursor_row)
        row=contents['values']
        self.Roll_No_var.set(row[0])
        self.name_var.set(row[1])
        self.email_var.set(row[2])
        self.rtype_var.set(row[3])
        self.contact_var.set(row[4])
        self.dob_var.set(row[5])
        self.txt_Address.delete("1.0",END)
        self.txt_Address.insert(END,row[6])

    def update_data(self):
        if self.Roll_No_var.get()=="" or self.name_var.get()=="" or self.email_var.get()=="" or self.rtype_var.get()=="" or self.contact_var.get()=="" or self.dob_var.get()=="":
            messagebox.showerror("Erroe","Nothing to update")

        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()
        cur.execute("update students set name=%s,email=%s,gender=%s,contact=%s,dob=%s,address=%s where roll_no=%s",(
                                                                                                self.name_var.get(),
                                                                                                self.email_var.get(),
                                                                                                self.rtype_var.get(),
                                                                                                self.contact_var.get(),
                                                                                                self.dob_var.get(),
                                                                                                self.txt_Address.get('1.0',END),
                                                                                                self.Roll_No_var.get()
                                                                                                ))
        con.commit()
        self.fetch_data()
        self.clear()
        con.close()

    def delete_data(self):
        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()
        cur.execute("delete from students where roll_no=%s",self.Roll_No_var.get())
        con.commit()
        con.close()
        self.fetch_data()
        self.clear()

    def search_data(self):
        if self.search_txt.get()=="":
            messagebox.showerror("Error","Enter somthing to search")
        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()

        cur.execute("select * from students where "+str(self.search_by.get())+" LIKE '%"+str(self.search_txt.get())+"%'" )
        rows=cur.fetchall()
        if len(rows)!=0:
            self.Student_table.delete(*self.Student_table.get_children())
            for row in rows:
                self.Student_table.insert('',END,values=row)
            con.commit()
        con.close()



wind1=Tk()
obj=Hostel(wind1)
wind1.mainloop()
```

```python
from tkinter import Tk, Label, Entry, Button, Frame, Text, messagebox
from tkinter import ttk
from PIL import Image, ImageTk
import pymysql

class Hostel:

    def __init__(self, wind1):
        self.wind1 = wind1
        self.wind1.title("Registration window")
        self.wind1.geometry("1350x700+0+0")
        # ----Background Image----#
        self.bkgrnd = ImageTk.PhotoImage(file="images/backgb.png")
        bkgrnd = Label(self.wind1, image=self.bkgrnd).place(x=0, y=0, relwidth=1, relheight=1)

        # ----Logo Image-----#
        self.logo = ImageTk.PhotoImage(file="images/1side.png")
        logo = Label(self.wind1, image=self.logo).place(x=80, y=100, width=400, height=500)

        # -----DAV LOGO TEXT-----#
        self.lbl1 = Label(self.wind1, text="DAV UNIVERSITY", font=("Bok Antiqua", 25, "bold"), fg="black",
                          compound='BOTTOM', bg="white")
        self.lbl1.place(x=90, y=100, height=50, width=350)

        frame1 = Frame(self.wind1, bg="white")
        frame1.place(x=480, y=100, width=700, height=500)

        # ----Login text----#
        title = Label(frame1, text="LOGIN HERE", font=("times new roman", 30, "bold"), bg="white", fg="#2744AB").place(x=80, y=50)

        # -----Login Email-----#
        self.l_email = Label(frame1, text="Email", font=("times new roman", 20, "bold"), bg="white", fg="gray")
        self.l_email.place(x=80, y=120)

        self.txt_l_email = Entry(frame1, font=("times new roman", 15), bg="lightgray")
        self.txt_l_email.place(x=80, y=160, width=400, height=25)

        # -----Login password-----#
        self.l_password = Label(frame1, text="Password", font=("times new roman", 20, "bold"), bg="white", fg="gray")
        self.l_password.place(x=80, y=220)

        self.txt_l_password = Entry(frame1, font=("times new roman", 15), bg="lightgray")
        self.txt_l_password.place(x=80, y=260, width=400, height=25)

        self.btn_reg = Button(frame1, text="Register New Account?", command=self.register_page, font=("times new roman", 18, "bold"), bg="white", bd=0, fg="#D42569", cursor="hand2")
        self.btn_reg.place(x=80, y=300)

        self.btn_login = Button(frame1, text="Login", font=("times new roman", 18, "bold"), command=self.login_data, bg="white", bd=0, fg="#D42569", cursor="hand2")
        self.btn_login.place(x=80, y=350)

    def register_page(self):
        self.wind1.destroy()
        self.wind2 = Tk()
        self.wind2.title("Login Window")
        self.wind2.geometry("1350x700+0+0")
        # ----Background Image-----#
        self.backgrnd = ImageTk.PhotoImage(file="images/backgb2.png")
        backgrnd = Label(self.wind2, image=self.backgrnd).place(x=0, y=0, relwidth=1, relheight=1)

        # ----Logo Image-----#
        self.logo = ImageTk.PhotoImage(file="images/1side.png")
        centr = Label(self.wind2, image=self.logo).place(x=80, y=100, width=400, height=500)

        # -----DAV LOGO TEXT-----#
        self.lbl2 = Label(self.wind2, text="DAV UNIVERSITY", font=("Bok Antiqua", 25, "bold"), fg="black",
                          compound='BOTTOM', bg="white")
        self.lbl2.place(x=90, y=100, height=50, width=350)

        # ----Register frame2-----#
        frame2 = Frame(self.wind2, bg="white")
        frame2.place(x=480, y=100, width=700, height=500)

        title = Label(frame2, text="REGISTER HERE", font=("times new roman", 20, "bold"), bg="white", fg="green").place(x=50, y=30)

        self.f_name = Label(frame2, text="First Name", font=("times new roman", 15, "bold"), bg="white", fg="gray")
        self.f_name.place(x=50, y=100)
        self.txt_fname = Entry(frame2, font=("times new roman", 15), bg="lightgray")
        self.txt_fname.place(x=50, y=130, width=250)

        self.l_name = Label(frame2, text="Last Name", font=("times new roman", 15, "bold"), bg="white", fg="gray")
        self.l_name.place(x=370, y=100)
        self.txt_lname = Entry(frame2, font=("times new roman", 15), bg='lightgray')
        self.txt_lname.place(x=370, y=130, width=250)

        # 2ROW
        self.contact = Label(frame2, text="Contact No.", font=("times new roman", 15, "bold"), bg="white", fg="gray")
        self.contact.place(x=50, y=170)
        self.txt_contact = Entry(frame2, font=("times new roman", 15), bg='lightgray')
        self.txt_contact.place(x=50, y=200, width=250)

        self.email = Label(frame2, text="Email", font=("times new roman", 15, "bold"), bg="white", fg="gray")
        self.email.place(x=370, y=170)
        self.txt_email = Entry(frame2, font=("times new roman", 15), bg='lightgray')
        self.txt_email.place(x=370, y=200, width=250)

        # 3ROW
        self.password = Label(frame2, text="Password", font=("times new roman", 15, "bold"), bg="white", fg="gray")
        self.password.place(x=50, y=240)
        self.txt_password = Entry(frame2, font=("times new roman", 15), bg='lightgray')
        self.txt_password.place(x=50, y=270, width=250)

        self.cpassword = Label(frame2, text="Confirm   Password", font=("times new roman", 15, "bold"), bg="white", fg="gray")
        self.cpassword.place(x=370, y=240)
        self.txt_cpassword = Entry(frame2, font=("times new roman", 15), bg='lightgray')
        self.txt_cpassword.place(x=370, y=270, width=250)

        # 4ROW--terms--
        self.btn_img1 = ImageTk.PhotoImage(file="images/rbutton.jpg")
        self.btn_register = Button(frame2, image=self.btn_img1, bd=0, cursor="hand2", command=self.register_data)
        self.btn_register.place(x=50, y=370)

        # self.btn_img2=ImageTk.PhotoImage(file="images/sbutton.png")
        # self.btn_signup=Button(self.wind2,image=self.btn_img2,bd=0,cursor="hand2")
        # self.btn_signup.place(x=180,y=500)

    def login_data(self):
        if self.txt_l_email.get() == "" or self.txt_l_password.get() == "":
            messagebox.showerror("Error", "All Fields Are Required", parent=self.wind1)
        else:
            try:
                con = pymysql.connect(host="localhost", user="root", password="", database="hostel")
                cur = con.cursor()
                cur.execute("select * from register where email=%s and password=%s", (self.txt_l_email.get(), self.txt_l_password.get()))
                row = cur.fetchone()
                if row is None:
                    messagebox.showerror("Error", "INVALID USER NAME AND PASSWORD", parent=self.wind1)
                else:
                    messagebox.showinfo("Success", "Welcome", parent=self.wind1)
                    self.wind1.destroy()
                    self.management()
                con.close()
            except Exception as es:
                messagebox.showerror("Error", f"Error due to: {str(es)}", parent=self.wind1)

    def management(self):
        self.wind3 = Tk()
        self.wind3.title("Hostel Management System")
        self.wind3.geometry("1350x700+0+0")

        # ----Background Image-----#
        self.bckgrnd = ImageTk.PhotoImage(file="images/backgb2.png")
        bckgrnd = Label(self.wind3, image=self.bckgrnd).place(x=0, y=0, relwidth=1, relheight=1)

        title = Label(self.wind3, text="North Hostel DAV University", bd=10, relief="groove", font=("times new roman", 40, "bold"), fg="black")
        title.pack(side="top", fill="x")

        # ----All Variables----#
        self.Roll_No_var = StringVar()
        self.name_var = StringVar()
        self.email_var = StringVar()
        self.rtype_var = StringVar()
        self.contact_var = StringVar()
        self.dob_var = StringVar()

        self.search_by = StringVar()
        self.search_txt = StringVar()

        # ---Manage Frame---#
        Manage_Frame = Frame(self.wind3, bd=4, relief="ridge")
        Manage_Frame.place(x=20, y=100, width=450, height=580)

        m_title = Label(Manage_Frame, text="Manage Students", font=("times new roman", 30, "bold"))
        m_title.grid(row=0, columnspan=2, pady=20)

        lbl_roll = Label(Manage_Frame, text="Roll No.", font=("times new roman", 20, "bold"))
        lbl_roll.grid(row=1, column=0, padx=20, pady=10, sticky="w")

        txt_Roll = Entry(Manage_Frame, textvariable=self.Roll_No_var, font=("times new roman", 15, "bold"), bd=5, relief="groove")
        txt_Roll.grid(row=1, column=1, padx=20, pady=10, sticky="w")

        lbl_name = Label(Manage_Frame, text="Name", font=("times new roman", 20, "bold"))
        lbl_name.grid(row=2, column=0, padx=20, pady=10, sticky="w")

        txt_Name = Entry(Manage_Frame, textvariable=self.name_var, font=("times new roman", 15, "bold"), bd=5, relief="groove")
        txt_Name.grid(row=2, column=1, padx=20, pady=10, sticky="w")

        lbl_email = Label(Manage_Frame, text="Email", font=("times new roman", 20, "bold"))
        lbl_email.grid(row=3, column=0, padx=20, pady=10, sticky="w")

        txt_Email = Entry(Manage_Frame, textvariable=self.email_var, font=("times new roman", 15, "bold"), bd=5, relief="groove")
        txt_Email.grid(row=3, column=1, padx=20, pady=10, sticky="w")

        lbl_rtype = Label(Manage_Frame, text="Room Type", font=("times new roman", 20, "bold"))
        lbl_rtype.grid(row=4, column=0, padx=20, pady=10, sticky="w")

        combo_type = ttk.Combobox(Manage_Frame, textvariable=self.rtype_var, font=("times new roman", 14, "bold"), state='readonly')
        combo_type['values'] = ("1 sitter non-A.C", "2 sitter non-A.C", "3 sitter non-A.C", "1 sitter A.C", "2 sitter A.C", "3 sitter A.C")
        combo_type.grid(row=4, column=1, padx=20,pady=10,sticky="w")
        self.txt_Address=Text(Manage_Frame,width=30,height=4,font=("",10))
        self.txt_Address.grid(row=7,column=1,padx=20,pady=10,sticky="w")

        #----button Frame----#
        btn_Frame=Frame(Manage_Frame,bd=4)
        btn_Frame.place(x=15,y=520,width=420)
        
        Add_btn=Button(btn_Frame,text="Add",width=10,command=self.add_students).grid(row=0,column=0,padx=10,pady=10)
        update_btn=Button(btn_Frame,text="Upadte",width=10,command=self.update_data).grid(row=0,column=1,padx=10,pady=10)
        delete_btn=Button(btn_Frame,text="Delete",width=10,command=self.delete_data).grid(row=0,column=2,padx=10,pady=10)
        clear_btn=Button(btn_Frame,text="Clear",width=10,command=self.clear).grid(row=0,column=3,padx=10,pady=10)

        #---Detail Frame----#
        Detail_Frame=Frame(self.wind3,bd=4,relief=RIDGE)
        Detail_Frame.place(x=500,y=100,width=800,height=580)

        lbl_Search=Label(Detail_Frame,text="Search By",font=("times new roman",20,"bold"))
        lbl_Search.grid(row=0,column=0,padx=20,pady=10,sticky="w")

        combo_search=ttk.Combobox(Detail_Frame,textvariable=self.search_by,width=10,font=("times new roman",14,"bold"),state='readonly')
        combo_search['values']=("Roll_no","Name","Contact")
        combo_search.grid(row=0,column=1,padx=20,pady=10,)

        txt_Search=Entry(Detail_Frame,textvariable=self.search_txt,width=20,font=("times new roman",10,"bold"),bd=5,relief=GROOVE)
        txt_Search.grid(row=0,column=2 ,padx=20,pady=10,sticky="w")

        search_btn=Button(Detail_Frame,text="Search",width=10,pady=5,command=self.search_data).grid(row=0,column=3,padx=10,pady=10)
        showall_btn=Button(Detail_Frame,text="Show All",width=10,pady=5,command=self.fetch_data).grid(row=0,column=4,padx=10,pady=10)

        #---Table frame----#
        Table_Frame=Frame(Detail_Frame,bd=4,relief=RIDGE)
        Table_Frame.place(x=10,y=70,width=760,height=500)

        scroll_x=Scrollbar(Table_Frame,orient=HORIZONTAL)
        scroll_y=Scrollbar(Table_Frame,orient=VERTICAL)
        self.Student_table=ttk.Treeview(Table_Frame,columns=("roll","name","email","room type","contact","dob","address"),xscrollcommand=scroll_x.set,yscrollcommand=scroll_y.set)
        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=RIGHT,fill=Y)
        scroll_x.config(command=self.Student_table.xview) 
        scroll_y.config(command=self.Student_table.yview)
        self.Student_table.heading("roll",text="Roll No.")
        self.Student_table.heading("name",text="Name")
        self.Student_table.heading("email",text="Email")
        self.Student_table.heading("room type",text="Room Type")
        self.Student_table.heading("contact",text="Contact")
        self.Student_table.heading("dob",text="D.O.B")
        self.Student_table.heading("address",text="Address")
        self.Student_table['show']='headings'
        self.Student_table.column("roll",width=100)
        self.Student_table.column("name",width=100)
        self.Student_table.column("email",width=100)
        self.Student_table.column("room type",width=100)
        self.Student_table.column("contact",width=100)
        self.Student_table.column("dob",width=100)
        self.Student_table.column("address",width=150)
        
        self.Student_table.pack(fill=BOTH,expand=1)
        self.Student_table.bind("<ButtonRelease-1>",self.get_cursor)

        self.fetch_data()
    
    
    def clear(self):
        self.txt_fname.delete(0,END)
        self.txt_lname.delete(0,END)
        self.txt_contact.delete(0,END)
        self.txt_email.delete(0,END)
        self.txt_password.delete(0,END)
        self.txt_cpassword.delete(0,END)


    def register_data(self):
        if self.txt_fname.get()=="" or self.txt_contact.get()=="" or self.txt_email.get()=="" or self.txt_password.get()=="" or self.txt_cpassword.get()=="":
            messagebox.showerror("Error","All Fields Are Required",parent=self.wind2)

        elif self.txt_password.get()!=self.txt_cpassword.get():
            messagebox.showerror("Error","Password & Confirm Password should be same",parent=self.wind2)

        else:
            try:
                con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
                cur=con.cursor()
                cur.execute("select * from register where email=%s",self.txt_email.get())
                row=cur.fetchone()
                if row!=None:
                    messagebox.showerror("Error","User already exist,Please try with another email",parent=self.wind2)
                else:
                    cur.execute("insert into register (f_name,l_name,contact,email,password) values(%s,%s,%s,%s,%s)",
                            (
                                self.txt_fname.get(),
                                self.txt_lname.get(),
                                self.txt_contact.get(),
                                self.txt_email.get(),
                                self.txt_password.get()
                            ))
                    con.commit()
                    con.close()
                    messagebox.showinfo("Success","Register Successfull",parent=self.wind2)
                    self.clear()
            except Exception as es:
                 messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.wind2)


    def login_data(self):
        if self.txt_l_email.get()=="" or self.txt_l_password.get()=="":
            messagebox.showerror("Error","All Fields Are Required",parent=self.wind1)

        else:
            try:
                con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
                cur=con.cursor()
                cur.execute("select * from register where email=%s and password=%s",(self.txt_l_email.get(),self.txt_l_password.get()))
                row=cur.fetchone()
                if row==None:
                    messagebox.showerror("Error","INVALID USER NAME AND PASSWORD",parent=self.wind1)
                else:
                    messagebox.showinfo("Success","Welcome",parent=self.wind1)
            except Exception as es:
                 messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.wind1)


    def add_students(self):
        if self.Roll_No_var.get()=="" or self.name_var.get()=="" or self.email_var.get()=="" or self.rtype_var.get()=="" or self.contact_var.get()=="" or self.dob_var.get()=="":
            messagebox.showerror("Erroe","All fields are required!!!",parent=self.wind3)
        
        else:
            con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
            cur=con.cursor()
            cur.execute("select * from students where email=%s",self.email_var.get())
            row=cur.fetchone()
            cur.execute("insert into students values(%s,%s,%s,%s,%s,%s,%s)",(self.Roll_No_var.get(),
                                                                         self.name_var.get(),
                                                                         self.email_var.get(),
                                                                         self.rtype_var.get(),
                                                                         self.contact_var.get(),
                                                                         self.dob_var.get(),
                                                                         self.txt_Address.get('1.0',END)
                                                                         ))
            con.commit()
            self.fetch_data()
            self.clear()
            con.close()
            messagebox.showinfo("Success","Record has been inserted")

    def fetch_data(self):
        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()
        cur.execute("select * from students")
        rows=cur.fetchall()
        if len(rows)!=0:
            self.Student_table.delete(*self.Student_table.get_children())
            for row in rows:
                self.Student_table.insert('',END,values=row)
            con.commit()
        con.close()

    def clear(self):
        self.Roll_No_var.set("")
        self.name_var.set("")
        self.email_var.set("")
        self.rtype_var.set("")
        self.contact_var.set("")
        self.dob_var.set("")
        self.txt_Address.delete("1.0",END)

    def get_cursor(self,ev):
        cursor_row=self.Student_table.focus()
        contents=self.Student_table.item(cursor_row)
        row=contents['values']
        self.Roll_No_var.set(row[0])
        self.name_var.set(row[1])
        self.email_var.set(row[2])
        self.rtype_var.set(row[3])
        self.contact_var.set(row[4])
        self.dob_var.set(row[5])
        self.txt_Address.delete("1.0",END)
        self.txt_Address.insert(END,row[6])

    def update_data(self):
        if self.Roll_No_var.get()=="" or self.name_var.get()=="" or self.email_var.get()=="" or self.rtype_var.get()=="" or self.contact_var.get()=="" or self.dob_var.get()=="":
            messagebox.showerror("Erroe","Nothing to update")

        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()
        cur.execute("update students set name=%s,email=%s,gender=%s,contact=%s,dob=%s,address=%s where roll_no=%s",(
                                                                                                self.name_var.get(),
                                                                                                self.email_var.get(),
                                                                                                self.rtype_var.get(),
                                                                                                self.contact_var.get(),
                                                                                                self.dob_var.get(),
                                                                                                self.txt_Address.get('1.0',END),
                                                                                                self.Roll_No_var.get()
                                                                                                ))
        con.commit()
        self.fetch_data()
        self.clear()
        con.close()

    def delete_data(self):
        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()
        cur.execute("delete from students where roll_no=%s",self.Roll_No_var.get())
        con.commit()
        con.close()
        self.fetch_data()
        self.clear()

    def search_data(self):
        if self.search_txt.get()=="":
            messagebox.showerror("Error","Enter somthing to search")
        con=pymysql.connect(host="localhost",user="root",password="",database="hostel")
        cur=con.cursor()

        cur.execute("select * from students where "+str(self.search_by.get())+" LIKE '%"+str(self.search_txt.get())+"%'" )
        rows=cur.fetchall()
        if len(rows)!=0:
            self.Student_table.delete(*self.Student_table.get_children())
            for row in rows:
                self.Student_table.insert('',END,values=row)
            con.commit()
        con.close()



wind1=Tk()
obj=Hostel(wind1)
wind1.mainloop()
```
